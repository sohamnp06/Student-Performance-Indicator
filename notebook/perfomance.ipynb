```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

DATA COLLECTION

```python
data = pd.read_csv("data/stud.csv")
data.head()
```

```python
print(data.info())
```

```python
# print("DUPLICATE VALUES = ",data.duplicated().sum())
# print(data.isnull().sum())
for col in data.columns:
    print(col, ":" ,data[col].unique(),"\nTotal : ",data[col].nunique())
```

--------------------------Insights-------------------------




1.DATASET HAS 1000 ROWS AND 8 COLUMNS

2.DATASET HAS NO NULL VALUES AND DUPLICATE VALUES


(3)   5- CATEGORICAL FEATURES AND 3- NUMERICAL FEATURES

```python
categorical_features=[col for col in data.columns if data[col].dtype == "O"]
print(categorical_features)
numerical_features=[col for col in data.columns if col not in categorical_features]
print(numerical_features)
```

```python
data.describe()
```

```python
data['total_score']=data['math_score']+data['reading_score']+data['writing_score']
data['average_score']=(data['total_score']/3).round(2)
data.head()
```

```python

data.to_csv("data\stud.csv")
```

```python
fig,ax=plt.subplots(2,3,figsize=(15,10))

sns.histplot(data=data, x='average_score', bins=30, ax=ax[0,0])
sns.histplot(data=data, x='average_score', bins=30, kde=True, hue='gender', ax=ax[0,1])
sns.barplot(data=data,x='gender',y='average_score',ax=ax[0,2])
sns.boxplot(data=data,y='math_score', ax=ax[1,0])
sns.boxplot(data=data,y='reading_score', ax=ax[1,1])
sns.boxplot(data=data,y='writing_score', ax=ax[1,2])

plt.tight_layout()
plt.show()
```

```python



plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('MATH SCORES')
sns.violinplot(y='math_score',data=data,color='red',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('READING SCORES')
sns.violinplot(y='reading_score',data=data,color='green',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('WRITING SCORES')
sns.violinplot(y='writing_score',data=data,color='blue',linewidth=3)
plt.show()
```
